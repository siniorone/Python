# Assertions in Python
```W3SCHOOL.com:``` The ```assert``` keyword is used when debugging code.
The ```assert``` keyword lets you test if a condition in your code returns True, if not, the program will raise an ```AssertionError```.
You can write a message to be written if the code returns False, check the example below.

````python
x = "hello sina"

#if condition returns False, AssertionError is raised:
assert x == "goodbye", "x should be 'hello sina'"
````

````python
def apply_discount(product, discount):
    price = int(product[ 'price' ] * ( 1.0 - discount))
    assert 0 <= price <= product[ 'price' ] , "Discount rate is ilegal"
    return price

graphic_card = { 'name' : 'Nvidia GeForce' , 'price' : 225}
apply_discount(graphic_card, discount=0.25)
````

let’s try to apply some invalid dis-
counts. For example, a **125%** “discount” that would lead to us giving
money to the customer:

````python
apply_discount(graphic_card, discount=1.25)
````

## The Most Commonly Python Assert  mistakes 
### 1.Don’t Use Asserts for Data Validation
assertions can be globally disabled with the and command line switches, as
well as the PYTHONOPTIMIZE environment variable in CPython.

````python
def delete_product(prod_id, user):
    assert user.is_admin(), 'Must be admin'
    assert store.has_product(prod_id), 'Unknown product'
    store.get_product(prod_id).delete()
````

If assertions are disabled in the Python interpreter, this turns into a null-op. Therefore any user can now delete products .

````python
def delete_product(product_id, user):
    if not user.is_admin():
        raise AuthError( 'Must be admin to delete' )
    if not store.has_product(product_id):
        raise ValueError ( 'Unknown product id' )
    store.get_product(product_id).delete()
````

### 2.Asserts That Never Fail
When a ```tuple``` is used as the first parameter in an ```assert``` statement, the assertion is always ```true```. 

````python
assert ( 1 == 2 , 'This should fail' )
````
